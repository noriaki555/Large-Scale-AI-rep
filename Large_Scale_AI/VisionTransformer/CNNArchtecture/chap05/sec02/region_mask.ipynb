```python
import torch

# 形状を定義して(1, 16, 16, 1)のテンソルを作成
shape = (16, 16)
region_mask = torch.zeros(1, *shape, 1)

window_size = 4
shift_size = 2

slices = [
    slice(0, -window_size),           # slice(0, -4)
    slice(-window_size, -shift_size), # slice(-4, -2)
    slice(-shift_size, None)          # slice(-2, None)
]

region_num = 0
for i in slices:
    for j in slices:
        region_mask[:, i, j, :] = region_num
        region_num += 1

print(region_mask.squeeze()) # 要素数が1の次元を削除して出力
```
