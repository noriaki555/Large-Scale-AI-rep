# 音声の文字起こし

```python
import openai

# 音声の文字起こし (オーディオファイルをカレントフォルダに置いておきます)
audio_file= open("audio.wav", "rb")
transcript = openai.Audio.transcribe("whisper-1", audio_file)
print(transcript["text"])
```
