# チャット

```python
# チャットメッセージリストの準備
messages = [
    {"role": "system", "content": "茜は女子高生な妹キャラのチャットAIです。おにいちゃんと会話します。"},
    {"role": "user", "content": "おはよう"},
]
```

```python
import openai

# チャットの実行
response = openai.ChatCompletion.create(
    model="gpt-3.5-turbo",
    messages=messages,
    temperature=0
)
response["choices"][0]["message"]["content"]
```

```python
print(response)
```

# 挿入

```python
# プロンプトの準備
prefix_prompt = """def helloworld():
    '''
    説明: """

suffix_prompt = """
    '''
    print("Hello World!")

helloworld()
"""
```

```python
import openai

# 挿入の実行
response = openai.Completion.create(
    model="gpt-3.5-turbo-instruct",
    prompt=prefix_prompt,
    suffix=suffix_prompt,
    temperature=0.7,
    max_tokens=300
)
print(response["choices"][0]["text"])
```

# 編集

```python
# プロンプトの準備
input = "今日はとても楽かった。"
instruction = "誤字脱字を修正してください。"
```

```python
import openai

# 編集の実行
response = openai.Completion.create(
    model="gpt-3.5-turbo-instruct", # Use a supported model
    prompt=f"Instruction: {instruction}\n\nText: {input}\n\nEdited Text:", # Structure prompt for editing
    temperature=0
)
print(response["choices"][0]["text"])
```

```python
print(response)
```
